```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 101
speed = np.linspace(0,1,n)

def speed_transfer(alpha=0.1,beta = 0.1,zeta=1):
    transformed_speed = (1-alpha-beta)*speed**3 + alpha*speed**2 + beta*speed
    speed_transfer = transformed_speed*(zeta-transformed_speed)
    return speed_transfer

def original():
    return speed**2 *(1-speed**2)
```

```python
aa,bb,zz = np.meshgrid(np.linspace(1,1,10),np.linspace(0,1,10),np.linspace(0.5,1.5,10))


for a,b,z in zip(aa.flatten(),bb.flatten(),zz.flatten()):
    otf = original()
    transfer = speed_transfer(a,b,z)
    plt.plot(speed,otf,'k')
    plt.plot(speed,transfer,'r--')
```

```python
aa,bb = np.meshgrid(np.linspace(0,1,10),np.linspace(0,0,10))


for a,b in zip(aa.flatten(),bb.flatten()):
    otf = original()
    transfer = speed_transfer(a,b)
    plt.plot(speed,otf,'k')
    plt.plot(speed,transfer,'r--')
```
